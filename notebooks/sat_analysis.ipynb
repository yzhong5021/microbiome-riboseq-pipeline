# Saturation Analysis of Metaribo-seq Library

This notebook visualizes the results of a saturation analysis I performed on the lab's ribo-seq library. I determined that the current size would not adequately capture full library complexity due to steadily rising number of unique reads as well as an extremely low duplicate read proportion.

***

```python
import matplotlib.pyplot as plt
import pandas as pd
```

```python
f = r'C:\Users\yoche\OneDrive - The University of Texas at Austin\Documents\Cenik\metaribo-seq\raw_data\saturation_analysis_results.tsv'
sat_df = pd.read_csv(f, sep = '\t', header = 0)
```

```python
x = sat_df['reads_in_subset']

fig, ax1 = plt.subplots()
ax1.set_xlabel('Total reads')
ax1.set_ylabel('Duplicate read proportion', color='blue')
ax1.plot(x, sat_df['saturation_score'], color='blue')
ax1.tick_params(axis='y', labelcolor='blue')

ax2 = ax1.twinx()
ax2.set_ylabel('Unique reads', color='red')
ax2.plot(x, sat_df['n_unique_sites'], color='red')
ax2.tick_params(axis='y', labelcolor='red')

fig.tight_layout()
plt.savefig('figures/sat_analysis.png', dpi=300)
plt.show()
```
